```python
import pandas as pd
```

```python
atp_2009 = pd.read_csv('data/atp_matches_2009.csv')
atp_2010 = pd.read_csv('data/atp_matches_2010.csv')
atp_2011 = pd.read_csv('data/atp_matches_2011.csv')
atp_2012 = pd.read_csv('data/atp_matches_2012.csv')
atp_2013 = pd.read_csv('data/atp_matches_2013.csv')
atp_2014 = pd.read_csv('data/atp_matches_2014.csv')
atp_2015 = pd.read_csv('data/atp_matches_2015.csv')
atp_2016 = pd.read_csv('data/atp_matches_2016.csv')
atp_2017 = pd.read_csv('data/atp_matches_2017.csv')
atp_2018 = pd.read_csv('data/atp_matches_2018.csv')
```

```python
list_of_df = [atp_2009, atp_2010, atp_2011, atp_2012, atp_2013, atp_2014, atp_2015, atp_2016, atp_2017, atp_2018]
```

```python
for i in list_of_df:
    if i.columns.all() == atp_2009.columns.all():
        print ('great')
    else:
        print ('not great')
```

```python
atp_10_years = pd.concat(list_of_df)
```

```python
atp_10_years = atp_10_years.reset_index().iloc[:,1:]
```

```python
atp_10_years.info()
```

```python
cols_to_keep = [
#     'tourney_id',
    'tourney_name',
    'surface',
    'draw_size',
    'tourney_level',
    'tourney_date',
#     'match_num',
    'winner_id',
    'winner_seed',
    'winner_entry',
    'winner_name',
    'winner_hand',
    'winner_ht',
    'winner_ioc',
    'winner_age',
    'winner_rank',
    'winner_rank_points',
    'loser_id',
    'loser_seed',
    'loser_entry',
    'loser_name',
    'loser_hand',
    'loser_ht',
    'loser_ioc',
    'loser_age',
    'loser_rank',
    'loser_rank_points',
#     'score',
    'best_of',
    'round']
```

```python
atp_10_years_red = atp_10_years[cols_to_keep]
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test = train_test_split(atp_10_years_red, test_size=0.5, random_state=0)
```

```python
X_train2 = X_train.rename(index=str, columns={"winner_id": "player1_id", 
                                            "winner_seed": "player1_seed",
                                            "winner_entry": "player1_entry",
                                            "winner_name": "player1_name",
                                            "winner_hand": "player1_hand",
                                            "winner_ht": "player1_ht",
                                            "winner_ioc": "player1_ioc",
                                            "winner_age": "player1_age",
                                            "winner_rank": "player1_rank",
                                            "winner_rank_points": "player1_rank_points",
                                            "loser_id": "player2_id", 
                                            "loser_seed": "player2_seed",
                                            "loser_entry": "player2_entry",
                                            "loser_name": "player2_name",
                                            "loser_hand": "player2_hand",
                                            "loser_ht": "player2_ht",
                                            "loser_ioc": "player2_ioc",
                                            "loser_age": "player2_age",
                                            "loser_rank": "player2_rank",
                                            "loser_rank_points": "player2_rank_points",
                                           })
```

```python
X_test2 = X_test.rename(index=str, columns={"winner_id": "player2_id", 
                                            "winner_seed": "player2_seed",
                                            "winner_entry": "player2_entry",
                                            "winner_name": "player2_name",
                                            "winner_hand": "player2_hand",
                                            "winner_ht": "player2_ht",
                                            "winner_ioc": "player2_ioc",
                                            "winner_age": "player2_age",
                                            "winner_rank": "player2_rank",
                                            "winner_rank_points": "player2_rank_points",
                                            "loser_id": "player1_id", 
                                            "loser_seed": "player1_seed",
                                            "loser_entry": "player1_entry",
                                            "loser_name": "player1_name",
                                            "loser_hand": "player1_hand",
                                            "loser_ht": "player1_ht",
                                            "loser_ioc": "player1_ioc",
                                            "loser_age": "player1_age",
                                            "loser_rank": "player1_rank",
                                            "loser_rank_points": "player1_rank_points",
                                           })
```

```python
X_train2['winner'] = 1
X_test2['winner'] = 0
```

```python
new = pd.concat([X_train2, X_test2])
```
